```python
from cgkit import graph
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

First we create the `GraphStructure`. In this case we set the biases to 0 everywhere, and draw weights as random integers between 0 and 10.

```python
G = graph.GraphStructure(continous = 5,
                         dummies = 1,
                         density = .3, 
                         seed = 3)

G.bias_space = lambda: 0
G.weight_space = lambda: np.random.randint(0,10)

G.draw_graph()
```

First we draw $1000\times 3$ samples and plot them all in a pairplot to check that the data generating process looks o.k.

```python
S = graph.CausalSystem(G, 3, 1000)
data = S.step_all()
```

```python
sns.pairplot(data, hue = 'idx', vars = data.columns[:-1])
```

Now lets pull out the true weights for each edge in the model. These will come in handy later

```python
true_vals = graph.find_true_values(G)
true_vals
```

We will also estimate a OLS regression pooled on this dataset just to check how well we're doing:

```python
import statsmodels.api as sm
ols = sm.OLS(data['c0'], data.drop(['c0', 'idx'], axis = 1))
ols_result = ols.fit()

ols_result.params
```

Finally we try to vary the sample size and study how well OLS does at different sample sizes. Throughout we use the same `GraphStructure` called `G`, but update the data generating system to alter the number of variables

```python
colors = {'d': 'orange',
         '0': 'purple',
         '1': 'red',
         '2': 'green',
         '3': 'blue',
         '4': 'C',
         '5': 'gray',
         '6':'tan',
         '7': 'black',
         '8': 'navy',
         '9':'aqua'}
```

```python
sizes = range(10,2000,20)
outdata = {i:list() for i in ols_result.params.keys()}

upper = {i:list() for i in ols_result.params.keys()}
lower = {i:list() for i in ols_result.params.keys()}

for size in sizes:
    g = graph.CausalSystem(G, 1, size)
    data = g.step_all()
    
    ols = sm.OLS(data['c0'], data.drop(['c0', 'idx'], axis = 1))
    ols_result = ols.fit()   
    
    res = ols_result.params
    std = ols_result.HC0_se
    
    for r in res.keys():
        outdata[r].append(res[r])
        upper[r].append(res[r] + 1.96*std[r])
        lower[r].append(res[r] - 1.96*std[r])
```

```python
fig, ax = plt.subplots(figsize=(15,10))

for var in sorted(outdata.keys()):
    ax.plot(sizes, outdata[var], color = colors[var[1]], alpha = 0.3, label = var)
    ax.fill_between(sizes, upper[var], lower[var],
    alpha=0.2, edgecolor=colors[var[1]], facecolor=colors[var[1]])

    
for t in true_vals:
    ax.plot(sizes, [true_vals[t] for _ in sizes], color = colors[t[1]], linestyle=':')
    
#ax.set_xscale("log", nonposx='clip')    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),
          fancybox=False, shadow=False, ncol=5)

ax.set_xlabel("Number of observations in data")
ax.set_ylabel("Parameter estimate")
```
