```python
from cgkit import *
from cgkit import graph2
import matplotlib.pyplot as plt 
import networkx as nx
import numpy as np
import pandas as pd 
import random
import seaborn as sns

import pretty
```

```python
s = graph2.GraphStructure(5,1,density = 1, seed = 4)
s.bias_space = lambda: 0
s.weight_space = lambda: np.random.randint(0,10)
s.draw_graph()
```

```python
g = graph2.CausalGraph(s, 1, 1000)
data = g.step_all()
```

```python
sns.pairplot(data, hue = 'idx', vars = data.columns[:-1])
```

```python
true_vals = dict()

for edge in s.edges:
    print(edge.i, edge.o)
    if edge.o[1] == 'y':
        true_vals[edge.i] = edge.weight
    else:
        if edge.i in true_vals:
            continue
        true_vals[edge.i] = 0

true_vals
```

```python
import statsmodels.api as sm
ols = sm.OLS(data['cy'], data.drop(['cy', 'idx'], axis = 1))
ols_result = ols.fit()

ols_result.params
```

```python
colors = {'d': 'orange',
         '0': 'purple',
         '1': 'red',
         '2': 'green',
         '3': 'blue',
         '4': 'C',
         '5': 'gray',
         '6':'tan',
         '7': 'black',
         '8': 'navy',
         '9':'aqua'}
```

```python
sizes = range(10,5000,50)
outdata = {i:list() for i in ols_result.params.keys()}

upper = {i:list() for i in ols_result.params.keys()}
lower = {i:list() for i in ols_result.params.keys()}

for size in sizes:
    g = graph2.CausalGraph(s, 1, size)
    data = g.step_all()
    
    ols = sm.OLS(data['cy'], data.drop(['cy', 'idx'], axis = 1))
    ols_result = ols.fit()   
    
    res = ols_result.params
    std = ols_result.HC0_se
    
    for r in res.keys():
        outdata[r].append(res[r])
        upper[r].append(res[r] + 1.96*std[r])
        lower[r].append(res[r] - 1.96*std[r])
```

```python
fig, ax = plt.subplots(figsize=(15,10))

for var in sorted(outdata.keys()):
    ax.plot(sizes, outdata[var], color = colors[var[1]], alpha = 0.3, label = var)
    ax.fill_between(sizes, upper[var], lower[var],
    alpha=0.2, edgecolor=colors[var[1]], facecolor=colors[var[1]])

    
for t in true_vals:
    ax.plot(sizes, [true_vals[t] for _ in sizes], color = colors[t[1]], linestyle=':')
#ax.set_xscale("log", nonposx='clip')    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=False, shadow=False, ncol=5)
```
